```python
import matplotlib.pyplot as plt
import numpy as np
```

There are 4 kinds of graphs in the Guerrero 2017 paper:
a)plot of the mean,min,minSOV values of the objective over the generations
b)plot of mean,min number of nodes and number of services over the generations
c)scatter plots of the pareto optimal front for any one objective vs another at some generation
d)obj values for different configurations for Kubernetes vs NSGA 

```python
def obj_over_gen(x_gens,y_obj_mean,y_obj_min,y_obj_minSOV,obj_name,title):
    plt.plot(x_gens,y_obj_mean,label='Mean')
    plt.plot(x_gens,y_obj_min,label='Min')
    plt.plot(x_gens,y_obj_minSOV,label='MinSOV')
    plt.xlabel('Generations')
    plt.ylabel(obj_name)
    plt.title(title)
    plt.legend()
    plt.show()
#filler data
gens = []
obj_vals_mean = []
obj_vals_min = []
obj_vals_minSOV = []
for i in range(1,30):
    gens.append(10*i)
    obj_vals_mean.append(np.random.randint(0,100))
    obj_vals_min.append(np.random.randint(0,100))
    obj_vals_minSOV.append(np.random.randint(0,100))
obj_over_gen(gens,obj_vals_mean,obj_vals_min,obj_vals_minSOV,"Objective name","Some title")
```

```python
def nodes_over_gen(x_gens,nodes_mean,nodes_min,title):
    plt.plot(x_gens,nodes_mean,label="Mean")
    plt.plot(x_gens,nodes_min,label="Min")
    plt.xlabel('Generations')
    plt.ylabel('Number of Nodes')
    plt.title(title)
    plt.legend()
    plt.show()
#filler data
gens = []
obj_vals_mean = []
obj_vals_min = []
for i in range(1,30):
    gens.append(10*i)
    obj_vals_mean.append(np.random.randint(0,100))
    obj_vals_min.append(np.random.randint(0,100))
nodes_over_gen(gens,obj_vals_mean,obj_vals_min,"Some title")
```

```python
def services_over_gen(x_gens,services_mean,services_min,title):
    plt.plot(x_gens,services_mean,label="Mean")
    plt.plot(x_gens,services_min,label="Min")
    plt.xlabel('Generations')
    plt.ylabel('Number of Services')
    plt.title(title)
    plt.legend()
    plt.show()
#filler data
gens = []
obj_vals_mean = []
obj_vals_min = []
for i in range(1,30):
    gens.append(10*i)
    obj_vals_mean.append(np.random.randint(0,100))
    obj_vals_min.append(np.random.randint(0,100))
services_over_gen(gens,obj_vals_mean,obj_vals_min,"Some title")
```

```python
def optimal_front_at_gen(x_obj,y_obj,x_obj_name,y_obj_name,title):
    plt.scatter(x_obj,y_obj)
    plt.xlabel(x_obj_name)
    plt.ylabel(y_obj_name)
    plt.title(title)
    plt.show()
#filler data
a=[]
b=[]
for i in range(10):
    a.append(np.random.randint(0,100))
    b.append(np.random.randint(0,100))
optimal_front_at_gen(a,b,"Balanced Cluster","Network distance","Generation 40")
```

```python
def obj_over_configs(x_config_names,y_kub,y_nsga,obj_name,title):
    x_idx = np.arange(len(x_config_names))
    width = 0.3
    plt.bar(x_idx,y_kub,width=width,label="Kubernetes")
    plt.bar(x_idx+width,y_nsga,width=width,label="NSGA-II")
    plt.xlabel('Experiment Configurations')
    plt.ylabel(obj_name)
    plt.title(title)
    plt.legend()
    plt.show()
#filler data
a=['1req-1app','1req-1.5app','1.5req-1app','1.5req-1.5app','2req-2app']
kub=[]
nsga=[]
for i in range(5):
    kub.append(np.random.randint(10,50))
    nsga.append(np.random.randint(10,50))
obj_over_configs(a,kub,nsga,"network distance","250 machines")
```
